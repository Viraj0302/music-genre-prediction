```python
import pandas as pd
music_data= pd.read_csv('/content/music.csv')
music_data
```

```python
#CLEANING THE DATASET
#to seperate the data set into 2 diff. sets- input&output. Since the genre is the output which we will recieve at the end acc. to the gender&age. we seperate it or rather create an input table w/o output
X=music_data.drop(columns=['genre']) #by convention we use X to represent this dataset
X
```

```python
 #now for the output dataset we use y as a convention. hence,
 y=music_data['genre']
 y
```

```python
#building a model
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier() #an instance of the DTC/ object
model.fit(X,y)
predictions=model.predict([ [21,1],[22,0] ]) #for a random sample data
predictions
```

```python
from sklearn.model_selection import train_test_split
train_test_split(X,y,test_size=0.2) 
#this returns a tuple which can be unpacked into 4 variables
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
 #input sets       #output sets
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train) #passing only the training dataset instead of entire dataset
predictions=model.predict(X_test)# passing passing dataset which contains values for testing instead of passing random datasets like before
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,predictions)
score
```
